```python
# NO OUTLIERS EXIST
#CHECK FOR NA
df.isna().sum()
```

```python
df['btc_total_bitcoins'].fillna(value=df['btc_total_bitcoins'].mean(),inplace=True)
```

```python
df['btc_trade_volume'].fillna(value=df['btc_trade_volume'].mean(),inplace=True)
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
#covert categorical data to continous data
my_object=df['Date']
```

```python
my_object_cov=pd.get_dummies(my_object,drop_first=True)
```

```python
df=df.drop('Date',axis=1)
```

```python
fin_df=pd.concat([df,my_object_cov],axis=1)
```

```python
fin_df.head()
```

```python
#create model
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X=fin_df.drop('btc_market_price',axis=1)
y=df['btc_market_price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
model=AdaBoostRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
MSE=mean_squared_error(y_test,pred)
RMSE=np.sqrt(MSE)
```
